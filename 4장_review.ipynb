###5.income 변수의 95% 부트스트랩 신뢰구간을 구하여라. 


```python
from sklearn.utils import resample
from sklearn.linear_model import LinearRegression
import numpy as np

# X는 피처 데이터, y는 타겟 데이터입니다.
# 이 데이터는 2차원 배열로 가정합니다.
X = np.array(df['income'])
y = np.array(df['gamble'])

# 샘플링 횟수와 신뢰구간을 설정합니다.
n_iterations = 1000
conf_level = 95

# 부트스트랩 샘플을 생성합니다.
bootstrapped_samples = list()
for i in range(n_iterations):
    sample = resample(list(zip(X, y)), replace=True)
    X_sample = np.array([row[0] for row in sample])
    y_sample = np.array([row[1] for row in sample])
    bootstrapped_samples.append((X_sample, y_sample))

# 모든 부트스트랩 샘플에 대해 회귀 모델을 학습하고 회귀 계수를 추출합니다.
coefficients = list()
for X_sample, y_sample in bootstrapped_samples:
    model = LinearRegression()
    model.fit(X_sample.reshape(-1, 1), y_sample)
    coefficients.append(model.coef_[0])

# 추출된 회귀 계수의 95% 신뢰구간을 계산합니다.
lower_bound = np.percentile(coefficients, 2.5)
upper_bound = np.percentile(coefficients, 97.5)

print("income 변수의 95% 신뢰구간: [{}, {}]".format(lower_bound, upper_bound))
```

###7.6번에서 계산한 예측값의 95% 부트스트랩 신뢰구간과 95% 부트스트랩 예측구간을 계산하여라. 

```python
#예측구간
X = np.array([[0, 45, 4.6, 7]])
model = LinearRegression()
model.fit(df.drop('gamble', axis=1), df['gamble'])
y_pred = model.predict(X)
model = LinearRegression()
model.fit(df.drop('gamble', axis=1), df['gamble'])
y_pred = model.predict(X)
errors = df['gamble'] - model.predict(df.drop('gamble', axis=1))
bootstrapped_errors = resample(errors, replace=True, n_samples=1000).tolist()
y_pred_list = [y_pred] * 1000


bootstrapped_predictions = y_pred_list + bootstrapped_errors
lower_bound = np.percentile(bootstrapped_predictions, 2.5)
upper_bound = np.percentile(bootstrapped_predictions, 97.5)
print(f"95% 예측 구간: [{lower_bound:.2f}, {upper_bound:.2f}]")
```

```python
# 신뢰구간
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.utils import resample
from scipy import stats
# 데이터를 불러옵니다.

# 설명변수와 반응변수를 설정합니다.
X = df[['gender', 'grading', 'income', 'GPA']]
y = df['gamble']

# 회귀분석 모델을 적합합니다.
model = LinearRegression().fit(X, y)

# 예측값을 구합니다.
x_new = np.array([0, 45, 4.6, 7]).reshape(1, -1)
y_pred = model.predict(x_new)

# 예측값과 실제값의 차이(오차)를 계산합니다.
errors = y - model.predict(X)

# 부트스트랩 샘플을 생성하며 예측값의 범위를 계산합니다.
n_iterations = 1000
alpha = 0.05
y_pred_list = []
for i in range(n_iterations):
    # 부트스트랩 샘플을 생성합니다.
    bootstrapped_errors = resample(errors, replace=True, n_samples=len(errors))
    y_pred_boot = y_pred + bootstrapped_errors
    
    # 예측값을 저장합니다.
    y_pred_list.append(y_pred_boot)

# 부트스트랩 샘플로부터 예측값의 신뢰구간을 계산합니다.
sigma = np.std(y_pred_list)
t_value = stats.t.ppf(1-alpha/2, n_iterations -1)
lower_bound = y_pred - t_value*sigma
upper_bound = y_pred + t_value*sigma
```

```python
print(f"95% 신뢰구간: [{lower_bound}, {upper_bound}]")
```

###8.잔차의 정규성을 확인하여라. 


```python
def bootstrap_residual(X, y, n_iterations=1000, alpha=0.05):
  
    # 예측값과 오차를 추출합니다.
    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)
    errors = y - y_pred
    
    # 오차의 부트스트랩 샘플을 생성합니다.
    bootstrapped_errors = resample(errors, replace=True, n_samples=n_iterations)
    
    return bootstrapped_errors

X = np.array(df.drop('gamble', axis=1))
y = np.array(df['gamble'])

bootstrapped_errors = bootstrap_residual(X, y, n_iterations=1000, alpha=0.05)
```

```python
print(pd.DataFrame(bootstrapped_errors).hist())
```

### 9.잔차와 예측값 (fitted value) 의 상관계수를 계산하고 산점도를 작성하여라. 


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array(df.drop('gamble', axis=1))
y = np.array(df['gamble'])
# 예측값과 오차를 추출합니다.
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
errors = y - y_pred
    
# 예측값과 잔차의 산점도를 작성합니다.
plt.scatter(y_pred, errors)
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
plt.title('Scatter plot of fitted values and residuals')

# 잔차와 예측값의 상관계수를 계산합니다.
correlation = np.corrcoef(y_pred, errors)[0, 1]
print('Correlation between residuals and fitted values:', correlation)


```

###10.income 변수에 대한 편잔차 그림(partial residual plot)를 작성하여라. 


```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 데이터를 불러옵니다.
df = pd.read_csv('/content/drive/MyDrive/2023년 4학년 2학기/데이터과학을 위한 통계/gamble.csv')

X = np.array(df['income'])
y = np.array(df['gamble'])

model = LinearRegression()
model.fit(X.reshape(-1, 1), y)
y_pred = model.predict(X.reshape(-1, 1))

partial_resid = y - y_pred
df = pd.DataFrame({'income': df['income'], 'Terms': y_pred, 'PartialResid': partial_resid})
plt.scatter(x=df['income'], y=df['PartialResid'], marker='o', edgecolors='black', facecolors='none')
plt.plot(df['income'], df['Terms'], linestyle='--')
plt.title('Partial Residual Plot for Income')
plt.xlabel('Income')
plt.ylabel('Partial Residuals')
plt.show()
```
